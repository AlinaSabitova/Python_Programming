Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель).

```python
def gcd_fun(a, b):
  if a > b:
    x = b
  else:
    x = a
  for i in range(1,a+1):
    if (a % i == 0) and (b % i == 0 ):
      gcd_fun = i
  return gcd_fun

def cal_lcm(a, b):
  if a > b:
    y = a
  else:
    y = b
  while(True):
    if (y % a == 0) and (y % b == 0):
      return y
    y +=1

def gcd_nums(nums1):
  for j in range(1, len(nums1)-1):
    nums1[j]=gcd_fun(nums1[j-1], nums1[j])
  return nums1[j] 

def lcm_nums(nums2):
  for j in range(1,len(nums2)):
    nums2[j]=cal_lcm(nums2[j-1], nums2[j])
  return nums2[j]
```

```python
nums1 = input("Введите числа через пробел: ").split()
nums1 = list(map(int,nums1))
nums2 = nums1.copy()
nod = gcd_nums(nums1)
nok = lcm_nums(nums2)
print("НОД=", nod)
print("НОК=", nok)

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:

отрицательных значений (отсортирован по убыванию);

неотрицательных значений (отсортирован по возрастанию).

```python
def split_numbers(*args):
  negative=[]
  positive=[]
  for i in args:
    if i >= 0:
      positive.append(i)
    elif i < 0:
      negative.append(i)
  return negative, positive
```

Дано натуральное число. Напишите рекурсивные функции для определения:

суммы цифр числа;

количества цифр в числе.

```python
def digits_sum(value):
  if value == 0:
    return 0
  else:
    return (value%10) + digits_sum(value//10) 



def digits_count(value):
  if value == 1:
    return 1
  else:
    return value**0 + digits_count(value//10)

print(digits_sum(12345))
print(digits_count(12345))

# --------------
# Пример вывода:
#
# 15
# 5
```
